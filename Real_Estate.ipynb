**Используя выводы из двух предыдущих заданий, найдите те наблюдения, где есть пропуски, выбросы (неадекватные значения признака, которые отличаются от большинства), используя фильтрацию из pandas.**

```python
outs = table.index[table['KitchenSquare'] > 1000].tolist()
outs += table.index[table['LifeSquare'] > 1000].tolist()
outs += table.index[table['HouseYear'] > 2021].tolist()
outs += table.index[table['Rooms'] == 0].tolist()
outs += table.index[table['Rooms'] == 19].tolist()
outs.sort()
df_outs = table.iloc[outs]
df_outs
```

```python
table.drop(table.index[outs], inplace = True)
filtered_stats = {
    "min" : table.min(axis=0),
    "max" : table.max(axis=0),
    "mean" : table.mean(axis=0)}
filtered_stats = pd.DataFrame(filtered_stats)
filtered_stats
```
